```python
import pyfirmata
from pyfirmata import Arduino, util, STRING_DATA
import word_gen
import random
import time

# https://gist.github.com/varlen/91170e7cdd61032107e833fce6b7106a

# https://pubmed.ncbi.nlm.nih.gov/28445740/


#ser = serial.Serial('/dev/cu.usbmodem142201', 9800, timeout=1)
#print(ser.readline())
board = pyfirmata.Arduino('/dev/cu.usbmodem142201') # this sometimes need to change. Check port.
board.send_sysex( STRING_DATA, util.str_to_two_byte_iter('starting up!') ) 

# Pins
left = 13
right = 12
amy = 11
mid = 10

emotions = ['sadness', 'anger', 'neutral', 'love', 'surprise', 'fear', 'happy', 'worry']
good_emotions = ['love', 'surprise', 'happy']
bad_emotions = ['sadness', 'anger', 'neutral', 'fear', 'worry']

it = pyfirmata.util.Iterator(board)
it.start()

#time.sleep(1)
sensor = board.get_pin('a:0:i')
mic = board.get_pin('a:1:i')

def msg( text ):
    if text:
        board.send_sysex( STRING_DATA, util.str_to_two_byte_iter( text ) )

time.sleep(3)

while True:
    distance = sensor.read() * 100

    hear = False

    # Need a range of time to detect sound
    for i in range(1000):
        volume = mic.read() * 1000
        if volume > 300:
            hear=True
            print("I hear you...")
            msg("I hear you...")
            break
    
    print("VOL: ", volume)
    print("DISTANCE: ", distance)

    if distance >70 and volume > 600:
        print("I feel connected to you.")
        msg("I feel connected to you.")
        board.digital[left].write(1)
        board.digital[right].write(1)
        board.digital[amy].write(1)
        board.digital[mid].write(1)

        time.sleep(5)
        print("Don't go.")
        msg("Don't go.")


    elif distance > 30 or hear:
        entry = random.choice(["Well hello", "See me", "Stop talking", "Be quiet!", "Don't tell me what to do", 
                               "I don't want to work", "Come closer...", "Let's talk!", "There you are!"])
        
        print(entry)
        while i < (len(entry)):
            msg(entry[i:i+16])
            time.sleep(1)
            i+=16
            msg(entry[i:i+16])
            time.sleep(2)
            i+=16

        emo = random.choice(good_emotions)
        say, theme = word_gen.theme_emotion_selector(emo)

        print("Emotion:")
        print(emo)
        
        
        msg("I feel ")
        time.sleep(1)
        msg(emo)
        time.sleep(2)
        msg("I'm thinking")
        msg(" about ")
        time.sleep(1)
        msg(theme)
        time.sleep(2)
        board.digital[right].write(1)

    else: 
        print("Isolated.")
        emo = random.choice(bad_emotions)
        say, theme = word_gen.theme_emotion_selector(emo)
        print("Emotion:")
        print(emo)
        if emo == 'sadness':
             board.digital[left].write(1)
        elif emo == 'anger' or emo == 'fear' or emo == 'worry':
            board.digital[amy].write(1)
        elif emo == 'neutral':
             board.digital[mid].write(1)


    print("Statement:")
    print(say)
    i = 0
    while i < (len(say)):
        msg(say[i:i+16])
        time.sleep(1)
        i+=16
        msg(say[i:i+16])
        time.sleep(2)
        i+=16
    
    board.digital[left].write(0)
    board.digital[right].write(0)
    board.digital[amy].write(0)
    board.digital[mid].write(0)

    
```
